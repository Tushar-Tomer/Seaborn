# relplot()/scatterplot()
## relpot takes bivariate/multivariate analysis
## When both the columns are numeric then we use relplot/scatterplot
## Both the plots are used in finding relation but relplot is much better than scatterplot. Both the plot use hue , size , style . But the first difference is relplot give the label outside the graph and scatterplot gives the label inside the graph . And second difference is relplot use the kind (kind gives the type of graph like line) but scatterplot not use the kind.
