```python
import pandas as pd
import seaborn as sns
```

```python
a=sns.load_dataset("tips")
a
```

```python
sns.relplot(x="total_bill",y="tip",data=a,kind="line");
```

```python
sns.relplot(x="total_bill",y="tip",data=a,hue="sex",kind="line");
```

```python
sns.relplot(x="total_bill",y="tip",data=a,hue="sex",style="smoker",kind="line");
```

```python
sns.lineplot(x="total_bill",y="tip",data=a,hue="sex");
```
