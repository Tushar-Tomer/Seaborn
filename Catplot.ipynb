```python
import seaborn as sns
```

```python
a=sns.load_dataset("tips")
a
```

# Find out average bill paid by male and female

```python
sns.catplot(x="sex",y="total_bill",data=a,kind="bar");
```

```python
sns.barplot(x="sex",y="total_bill",data=a);
```

```python
sns.catplot(x="sex",y="total_bill",data=a,kind="bar",estimator=sum);
```

# Average tip given by male and female

```python
sns.catplot(x="sex",y="tip",data=a,kind="bar");
```

```python
sns.catplot(x="sex",y="tip",data=a,kind="bar",hue="smoker");
```

# At what time maximum bill was paid

```python
sns.catplot(x="time",y="total_bill",data=a,kind="bar",estimator=max);
```

# Find out the hot selling day

```python
sns.catplot(x="day",y="total_bill",data=a,kind="bar",estimator=sum);
```
