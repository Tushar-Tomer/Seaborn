# Year wise passengers

```python
a=sns.load_dataset("flights")
a
```

```python
import matplotlib.pyplot as plt
plt.plot(a["year"],a["passengers"]);
plt.xlabel("Years");
plt.ylabel("Passengers");
```

```python
sns.relplot(x="year",y="passengers",data=a);
```

```python
sns.relplot(x="year",y="passengers",data=a,kind="line");
```

```python
sns.relplot(x="year",y="passengers",data=a,hue="month",kind="line");
```

```python
sns.relplot(x="month",y="passengers",data=a,kind="line");
```
